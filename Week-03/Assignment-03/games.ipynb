```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vg = pd.read_csv('vgsales.csv')
vg
```

cleaning

```python
Game_Values = vg.groupby('Name')['Global_Sales'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=Game_Values.index,y=Game_Values.values)
plt.xticks(rotation=90)
```

```python
vg = vg.dropna(subset=['Year'])
vg['Year'] = vg['Year'].astype(int)
```

```python
genre_sales = vg.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=genre_sales.index, y=genre_sales.values)
plt.title("Top 10 Game Genres by Global Sales")
plt.ylabel("Sales (Millions)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
platform_sales = vg.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=platform_sales.index, y=platform_sales.values)
plt.title("Top 10 Platforms by Global Sales")
plt.ylabel("Sales (Millions)")
plt.xticks
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(vg['Global_Sales'], bins=80, kde=True)
plt.title("Distribution of Global Game Sales")
plt.xlabel("Sales (Millions)")
plt.ylabel("Number of Games")
plt.tight_layout()
plt.show()
```

```python
sales_year = vg.groupby('Year')['Global_Sales'].sum()
plt.figure(figsize=(10, 5))
plt.plot(sales_year.index, sales_year.values, marker='o', color='green')
plt.title("Global Sales Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Sales (Millions)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
publisher_sales = vg.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=publisher_sales.index, y=publisher_sales.values)
plt.title("Top 10 Publishers by Global Sales")
plt.ylabel("Sales (Millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
label_counts = vg['Platform'].value_counts().head(20)
plt.figure(figsize=(8, 8))
plt.pie(label_counts, labels=label_counts.index, autopct='%2.1f%%', colors=sns.color_palette('viridis', len(label_counts)))
plt.title('Pie Chart of Platform Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
type_counts = vg['Genre'].value_counts().head(6)
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=sns.color_palette('husl', 6))
plt.title('Distribution of Top 6 Games Types')
plt.show()
```
